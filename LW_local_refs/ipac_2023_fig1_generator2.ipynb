```julia
using Plots, LaTeXStrings

# Define the parameters
e = 1.60217662e-19  # Electron charge in Coulombs
R = 1  # meters

# Define the range of beta values
beta = range(0.01, stop=0.9999, length=100)  # beta values ranging from 0 to 1
beta_2 = range(-0.01, stop=-0.9, length=100)

# Calculate the corresponding E_n values using the equation
E_n = e * ((1 .- beta.^2) ./ ((1 .- beta).^2 * R^2))
E_n_2 = e * ((1 .- beta_2.^2) ./ ((1 .- beta_2).^2 * R^2))

# Convert electric field to eV/m
E_n = E_n / e  # Convert from V/m to eV/m
E_n_2 = E_n_2 / e  # Convert from V/m to eV/m

# Create the plot with logarithmic scale on the y-axis
vline([0.99999],linestyle=:dash,label=:none,line=3,color=:black,alpha=0.45)
plot!(beta, E_n, xlabel=L"\gamma", ylabel=L"\textrm{E}_n~~\textrm{[eV}\,\,\textrm{m^{-1}]}", yscale=:log10, line=4.5,
     label=L"{\beta}>1", tickfontsize=10, guidefontsize=14,
     grid=0, legendfontsize=14, legend=:topleft,color=1)
plot!(abs.(beta_2), E_n_2, yscale=:log10, line=4.5, xscale=:log10, label=L"{\beta}<1",color=2,grid=false)
xticks!([0.1, 0.9999], ["1.15", "20"])
#yticks!([0.1, 1, 10^5], ["0","1",L"10^5"])

savefig("/Users/benfolsom/Documents/ipac_2023_fig1_v3.pdf")
```

```julia
using Plots, LaTeXStrings

# Define the parameters
e = (1.60217662e-19)^2  # Square of the electron charge in Coulombs^2
R = 1e-10  # Convert R from millimeters to meters

# Define the range of beta values
beta = range(0.2, stop=0.999999999999, length=100)  # beta values ranging from 0 to 1
beta_2 = range(-0.2, stop=-0.9, length=100)

# Calculate the corresponding E_n values using the equation
E_n = e^2 * ((1 .- beta.^2) ./ ((1 .- beta).^2 * R^2))
E_n_2 = e^2 * ((1 .- beta_2.^2) ./ ((1 .- beta_2).^2 * R^2))

# Convert electric field to N/m
E_n = E_n / e  # Convert from C^2/m to N/m
E_n_2 = E_n_2 / e  # Convert from C^2/m to N/m

# Create the plot with logarithmic scale on the y-axis
plot(beta, E_n, xlabel=L"\gamma", ylabel="Force (N/m)", yscale=:log10, line=4.5,
     label=L"\vec{\beta}>1", tickfontsize=10, guidefontsize=14,
     grid=0, legendfontsize=13, legend=:topleft)
plot!(abs.(beta_2), E_n_2, yscale=:log10, line=4.5, xscale=:log10, label=L"\vec{\beta}<1")
xticks!([0.2, 0.6, 1], [L"1.2", L"2.3", L"10^6"])
```

```julia
using Plots

# Define data
beta = 0.1
gamma = 1 / sqrt(1 - beta^2)
n = [1, 0, 0]
e = 1
R = 1

# Calculate E_n
E_n = e * ((1 - beta^2) / ((1 - beta)^2 * R^2)) * n

# Plot
plot([beta], [norm(E_n)], xlabel="β", ylabel="||E_n||", title="Electric Field", legend=false, lw=2, m=:o, ms=4, mc="red", yscale=:log10)
xlabel!("β")
ylabel!("||E_n||")

# Add Greek letters in bold to plot
annotate!(0.8, norm(E_n), text("E_n = $(beta)", :bold, :left, :top, :red))
annotate!(0.8, norm(E_n)/10, text("γ = $(gamma)", :bold, :left, :top, :black))
```
