```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def z_func(gamma, R):
    beta = np.sqrt((gamma**2 - 1) / gamma**2)
    return (1.178734E-5*(1 + beta) / ((1 - beta) * R**2))

# Define the range of beta values to plot
beta_vals_1 = np.linspace(0.01, 0.9999, 100)
beta_vals_2 = np.linspace(-0.6, -0.01, 100)

# Calculate the corresponding E_n values using the equation
E_n_1 = z_func(beta_vals_1, 1.0)
E_n_2 = z_func(beta_vals_2, 1.0)

# Create the plot
plt.figure()
plt.plot(beta_vals_1, E_n_1, label=r'$\vec{β}>1$', line_cap='round', linestyle='-', linewidth=4.5)
plt.plot(beta_vals_2, E_n_2, label=r'$\vec{β}<1$', line_cap='round', linestyle='-', linewidth=4.5)

plt.xlabel(r'$\beta$')
plt.ylabel(r'$\mathbf{E}_n ~ [V/m]$')

plt.yscale('log')
plt.xscale('log')

plt.xlim(0.01, 1.0)
plt.ylim(1e-12, 1e6)

plt.title(r'$\mathbf{E}_n ~ [V/m]$ for $R=1.0$')

plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Define the parameters
e = 1.178734E-5  # constant value for e
R = 0.01  # constant value for R

# Define the range of beta values
beta = np.linspace(0.04, 0.999, 100)  # beta values ranging from 0 to 1
beta_2 = np.linspace(-0.04, -0.999, 100)

# Calculate the corresponding E_n values using the equation
E_n = e * ((1 + beta) / ((1 - beta) * R**2))
E_n_2 = e * ((1 + beta_2) / ((1 - beta_2) * R**2))

# Create the plot with logarithmic scale on the y-axis
fig, ax = plt.subplots()
ax.plot(beta, E_n, label=r'$\mathbf{\beta} > 1$', linewidth=2.5, linestyle='-', color='blue')
ax.plot(np.abs(beta_2), E_n_2, label=r'$\mathbf{\beta} < 1$', linewidth=2.5, linestyle='--', color='red')

# Set the x-axis and y-axis to logarithmic scale
ax.set_xscale('log')
ax.set_yscale('log')
#ax.set_xlim(0,0.999999999999)
# Set labels with LaTeX rendering
ax.set_xlabel(r'$|\beta|$', fontsize=14)
ax.set_ylabel(r'$\mathbf{E}_n$ [V/m]', fontsize=14)

# Customize tick formatting for y-axis
ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
#ax.yaxis.set_major_locator(plt.LogLocator(base=10.0, numticks=15))  # Adjust the number of y-axis ticks
ax.set_xticks([0.1,0.999])
ax.set_ylim(0.0015,300)
# Add legend
ax.legend(fontsize=13, loc='upper left')

# Remove the grid
ax.grid(False)

# Show the plot
plt.tight_layout()
#plt.show()
plt.savefig('WEPL105_f1b.png')
```
