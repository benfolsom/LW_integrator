```python
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return (1 + x) / (1 - x)


x = np.linspace(-1, 1, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of (1+x)/(1-x) for -1 < x < 1")
plt.xlim(-0.9999, 0.9999)
# plt.ylim(-3, 3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-1, 1, 400)  # Generate 400 points between -1 and 1

# Calculate the corresponding y values for the function
y = (1 + x) / (1 - x)

# Create the plot
plt.plot(x, y, label="(1 + x) / (1 - x)", color="blue")

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of (1 + x) / (1 - x)")

# Add a grid
plt.grid()

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
